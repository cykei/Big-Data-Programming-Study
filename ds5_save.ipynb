## 문제 4: url에서 파일을 가져와 자료구조에 저장하기

```python
import urllib2
url='http://archive.ics.uci.edu/ml/machine-learning-databases/horse-colic/horse-colic.data'
res=urllib2.urlopen(url)
html=res.read()
res.close()
print len(html)
```

```python
lines=html.splitlines()
data=[]
for line in lines:
    data.append(line.split())
print len(data),len(data[0])
print data[0]
```

```python
# 데이터 확인1 - 모든 300행에 대한 28개 컬럼이 있는가
for i in range(0,len(data)):
    if len(data[i])!=len(data[0]):
        print "this columm has wrong size at" ,i
print "this data is perfect"
```

```python
# 데이터 확인2 - 처음 20개만 3번째 컬럼 평균내보기
sum=0;
cnt=0;
for i in range(0, 20):
    val= data[i][3]
    if val is '?':
        print i,"None"
    else:
        sum+=float(val)
        cnt+=1
        print i,val, sum
average=float(sum/cnt)
print "count={0} sum={1} average={2:2.2f}".format(cnt,sum,average)
```

```python
# 데이터 확인 개인 -물음표 갯수세기
cnt=0;
for i in range(0,len(data)):
    for j in range(0,len(data[0])):
        if data[i][j]=='?':
            cnt+=1;
            print "(",i,",",j,")",
    print "->",cnt
                
```

## 문제 5 : mongodb에 'id', '이름'을 저장하고, 조회하는 프로그램 실행하기

```python
from pymongo import MongoClient

client=MongoClient()
db=client.mypDB
```

```python
_id=1
_name='js'
_age=11
_country='ko'
```

```python
db.myPyCol.insert_one({
    "id": _id,
    "name": _name,
    "age": _age,
    "country": _country
})
```

```python
results = db.myPyCol.find()
for r in results:
        print r['name']
```

```python
_name='jslim'
_age=123
db.myPyCol.update_one(
        {"id": _id},
        {
            "$set": {
                "name":_name,
                "age":_age,
            }
        }
    )
```

```python
results = db.myPyCol.find()
for r in results:
        print r['id'],r['name']
```

```python
db.myPyCol.delete_many({"id":1})
```

```python
results = db.myPyCol.find()
for r in results:
        print r['id'],r['name']
```
